```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
shootings2014 = "data/2014shootings.csv"
shootings2014_data_df = pd.read_csv(shootings2014)

shootings2014_data_df = shootings2014_data_df[['Incident Date', 'State', 'City Or County', 'Address', '# Killed', '# Injured', 'Operations']]

shootings2014_data_df.head()
```

```python
shootings2015 = "data/2015shootings.csv"
shootings2015_data_df = pd.read_csv(shootings2015)

shootings2015_data_df = shootings2015_data_df[['Incident Date', 'State', 'City Or County', 'Address', '# Killed', '# Injured', 'Operations']]


shootings2015_data_df.head()
```

```python
shootings2016 = "data/2016shootings.csv"
shootings2016_data_df = pd.read_csv(shootings2016)
shootings2016_data_df.head()
```

```python
shootings2017 = "data/2017shootings.csv"
shootings2017_data_df = pd.read_csv(shootings2017)
shootings2017_data_df.head()
```

```python
shootings2018 = "data/2018shootings.csv"
shootings2018_data_df = pd.read_csv(shootings2018)
shootings2018_data_df.head()
```

```python
shootings2019 = "data/2019shootings.csv"
shootings2019_data_df = pd.read_csv(shootings2019)
shootings2019_data_df.head()
```

```python
shootings_combined = pd.concat([shootings2014_data_df, shootings2015_data_df, shootings2016_data_df, shootings2017_data_df, shootings2018_data_df, shootings2019_data_df])
shootings_combined = shootings_combined[['Incident Date', 'State', 'City Or County', 'Address', '# Killed', '# Injured']]
shootings_combined
```

```python
shootings_combined.to_csv('data/combined.csv')
```
